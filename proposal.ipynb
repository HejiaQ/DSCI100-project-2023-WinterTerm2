```python
import altair as alt
import numpy as np
import pandas as pd
```

```python
knowledge_data= pd.read_excel("data/user_model.xls", 2, usecols="A:F")
knowledge_data.columns=knowledge_data.columns.astype(str)
knowledge_data=knowledge_data.rename(columns={" UNS": "level"})
knowledge_data
```

```python
plot=(
    alt.Chart(knowledge_data)
    .mark_point()
    .encode(
        x=alt.X("LPR", title="study level"),
        y=alt.Y("STG", title= "performance"),
        color="level"
    )
)
plot
```
